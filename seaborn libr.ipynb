```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x=tips.sex) #it is useful to count no.of time each unique value is repeated
```

```python
tips.shape
```

```python
sns.distplot(x=tips['tip'],bins=5) # in seaborn library (distplot-to check how data distributed)
```

```python
sns.distplot(x=tips['tip'])
```

```python
#kernel density estimation plot
#better alternative to histogram as you can see data skewness
#kde showns line
sns.kdeplot(tips['tip'],shade=True)
```

```python
#shows the distribution of two parameter together
#jointplot
sns.jointplot(x=tips['total_bill'],y=tips['tip']) #here possitive corelaton can be seen
```

```python
#shows the regression of two parameter together
#jointplot
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde')
```

```python
sns.swarmplot(x='species',y='sepal_length',data=iris)
```

```python
#violin plot
sns.violinplot(x='sex',y='tip',data=tips,hue='time')
```

```python
#heatmap

#the heatmap is a way of representing the data in a 2 dimensional form
#the data values are representing as colours in the graph
#the goal of the heatmap is to provide a coloued visual summary of information

tips.corr()
```

```python
#cbar=colour bar like temperature scale om roght side 

#cmap=colormap(colour of boxes)      #note:should be mention different colors for better understanding

#annot=True means notation/value will be seen on each boxes

#linewidths=width of line in btwn boxes   #linecolor=line color of width

#yticklabels=values of y-axis    #xticklabels=values of x-axis

#center=0,1.....means intensity/transaparency of colour

import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
sns.heatmap(tips.corr(),yticklabels=True,
            cbar=True,cmap='Oranges',center=1,
            annot=True,linewidths=0,
            linecolor='red')
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(tips.isna(),yticklabels=True,cbar=True,cmap='YlGnBu',
           annot=True,
           linewidths=0,linecolor='grey')
```

```python
#box plot in seaborn

sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
#relation plot
#it uses extra keywords to use maximum data points in graph
```

```python
sns.relplot(x="total_bill",y="tip",col="time",
          hue="size",style="smoker",size="size",
          palette="YlGnBu",markers=["D", "o"],sizes=(10,125),
          edgecolor=".2",linewidth=.5,alpha=.75,data=tips)
```
